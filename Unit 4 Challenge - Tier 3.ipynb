#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Transpose the dataframe so dates become columns and boroughs become observations
Tr_properties = properties.transpose()
Tr_properties.columns = Tr_properties.iloc[0].values
Tr_properties = Tr_properties.drop(Tr_properties.index[0])
Tr_properties = Tr_properties.reset_index()
Tr_properties.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Rename columns for index and NaT
Tr_properties.rename(columns={'index':'London Borough', pd.NaT:'ID'}, inplace=True)
Tr_properties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melt the dataframe
tidy_properties = pd.melt(Tr_properties, id_vars = ['London Borough', 'ID'], var_name = 'Month', value_name = 'Average Price')
tidy_properties.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Check datatype of each column. 
tidy_properties.dtypes

# Three columns are of type 'object' meaning it's filled with mixed types. Let's change 
# each column accordingly.
tidy_properties['London Borough'] = tidy_properties['London Borough'].astype('string')
tidy_properties['ID'] = tidy_properties['ID'].astype('string')
tidy_properties['Average Price'] = tidy_properties['Average Price'].astype('float64')

# Check again for datatype of each column and you'll see appropraite data types. 
tidy_properties.dtypes
```

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit.

```python
# Let's drop the entire rows for missing values for 'Unnamed: 37'
unnamed_37 = tidy_properties['London Borough'] != 'Unnamed: 37'
tidy_properties = tidy_properties[unnamed_37]
```

```python
# Let's drop the entire rows for missing values for 'Unnamed: 47'
unnamed_47 = tidy_properties['London Borough'] != 'Unnamed: 47'
tidy_properties = tidy_properties[unnamed_47]
```

```python
### List of boroughs not in London
not_borough = ['NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 'EAST MIDLANDS', 'WEST MIDLANDS', 'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 'SOUTH WEST', 'England', 'City of London', 'Inner London', 'Outer London']

### Iterate through boroughs in list and remove from dataframe
for non_borough in not_borough:
    tidy_properties = tidy_properties[tidy_properties['London Borough'] != non_borough]
    
### Unique values in 'London Borough' column 
tidy_properties['London Borough'].unique()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
### Function takes dataframe and borough name and outputs price ratio between 1998 and 2018
def create_price_ratio(df, borough):
    df = df[df['London Borough']== borough]
    filter_1998 = df[df['Year']==1998] 
    avgPrice_1998 = filter_1998['Average Price'].mean()
    filter_2018 = df[df['Year']==2018]
    avgPrice_2018 = filter_2018['Average Price'].mean()
    return (avgPrice_1998)/(avgPrice_2018)

        # Create list of unique boroughs in London: borough_list
borough_list = tidy_properties['London Borough'].unique()

### Get price ratio between 1998/2018 for each borough by calling create_price_ratio function
### for each borough.
PRatio_Dict ={}   # Create empty dictionary to store price ratio for each borough
    # iterate through borough_list
for borough in borough_list:
    price_ratio = create_price_ratio(tidy_properties, borough)
    PRatio_Dict[borough] = price_ratio

### GREAT! We have our PRatio_Dict with boroughs and corresponding ratios. Let's put them 
### in increasing order based off the price ratios as list of tuples: sorted_PRatio 
import operator
sorted_PRatio = sorted(PRatio_Dict.items(), key=operator.itemgetter(1))
print('Below are the top five boroughs with largest increase in average home price between 1998 and 2018:')
print(sorted_PRatio[0:5]); print('\n')  # top 5 boroughs with largest increase in price over 2 decades

### The most expensive boroughs in 2020:
filter_2020 = tidy_properties[tidy_properties['Year']==2020]
LBorough_Average = filter_2020.groupby(['London Borough']).mean()

print('Boroughs of 2020, listed in order. The first borough is the most expensive:')
sorted_LBorough = LBorough_Average.sort_values('Average Price', ascending=False)
print(sorted_LBorough)
```
